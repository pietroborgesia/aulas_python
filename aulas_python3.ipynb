```python
#Verificar a aprovação ou não

nota_aluno = float(input("Digite a sua média final"))
nota_aluno
```

```python
if nota_aluno >= 7:
    print("Parabéns, você foi aprovade!")

elif nota_aluno >= 5:
    print("Você está em recuperação!")

elif nota_aluno < 0:
    print("Você tirou nota negativa? \n Parabéns")

else:
    print("Você está reprovado!")
```

```python
#Resto da divisão por um número '%'
#Par ou ímpar

num = int(input("Digite um número natural"))
if num%2 == 0:
        print(f"O {num} é par")
else:
    print(f"O {num} é ímpar!")
```

```python
#Comaprar dois números

num1 = float(input("Digite o primeiro número:"))
num2 = float(input("Digite o segundo número:"))

if num1 > num2:
    print(f"{num1} > {num2}")
elif num1 == num2:
    print(f"{num1} = {num2}")
else:
    print(f"{num2} > {num1}")
```

```python
#Resolvendo uma equação do segundo grau
# '**' é elevado
#break é para parar o programa

a = float(input("Digite o valor do coeficiente a"))
b = float(input("Digite o valor do coeficiente b"))
c = float(input("Digite o valor do coeficiente c"))
delta = b**2 - 4*a*c
x1 = (-b + delta**(0.5))/(2*a)
x2 = (-b - delta**(0.5))/(2*a)
if delta < 0:
    print(f"Delta menor do zero:{delta}")
    print("Não há raiz real")

elif a == 0:
    print(f"{a} é menor que zero, insira um número diferente de zero")
elif delta == 0:
    print(f" x'={x1} = x''={x2}")
else:
    print(f" x'={x1} = x''={x2}")
```
