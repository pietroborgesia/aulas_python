# Variável Local e Global

```python
#variável global
var_global = 10

def multiplica(num1, num2):
    var_global = num1 * num2
    print(var_global)
```

```python
multiplica(5,5)
```

```python
var_global
```

```python
# Variável Local
var_global = 10

def multiplica(num1, num2):
    local = num1 * num2
    return local
```

```python
multiplica(20,5)
```

```python
resultado = multiplica(20,5)
print(resultado)
print(var_global)
var_global
```

```python
def numprimo(num):
  if num % 2 == 0 and num > 2:
    return "Este número não é primo"
  for i in range(3, int(math.sqrt(num)) + 1, 2):
    if num % i == 0:
      return "Este número não é primo"
  return "Este número é primo"
```

```python
numprimo(89)
```

# Funções Lambda

```python
# Definindo uma função - 3 linhas de código
def potencia(num):
  resultado = num ** 2
  return resultado
```

```python
num = float(input("Digite um número que queira o quadrado: "))
potencia(num)
```

```python
# Definindo uma função - 2 linhas de código

def potencia2(num):
  return num**2
```

```python
potencia2(num)
```

```python
# Definindo uma função - 1 linha de código

def potencia3(num): return num**2
```

```python
potencia3(num)
```

```python
# Definindo uma expressão Lambda
potencia4 = lambda num: num**2
```

```python
potencia4(6)
```

```python
# Impar ou par lambda

par = lambda x: x%2 == 0
```

```python
par(35)
```

```python
impar = lambda x: x%2 != 0
```

```python
impar(35)
```

```python
#expressão para imprimir a primeira letra

first_letter = lambda s: s[0]
```

```python
first_letter("Pietro")
```

```python
cont_letras = lambda s: len(s)
```

```python
word = "Pietro"
print(f"A palavra '{word}' tem {cont_letras(word)} letras." )
```

```python
#expressão para imprimir a última letra

last_letter = lambda s: s[-1]
```

```python
last_letter("Pietro")
```

```python
#escrever a palavra de trás para frente

reverso = lambda s: s[ ::-1 ]
```

```python
reverso("Análise de Dados")
```
