## Exercício 1.

```python
#Crie uma função que imprima a sequência de números pares
#entre 1 e 20 (a função não recebe parâmetro) e depois faça uma chamada à
#função para listar os números.

def imprimir_pares():
    for i in range(1, 21):
        if i % 2 == 0:
            print(i)

# chamada da função
imprimir_pares()
```

## Exercício 2.

```python
def para_maiusculas(texto):
    return texto.upper()

# chamada da função
resultado = para_maiusculas("python funções")
print(resultado)
```

## Exercício 3.

```python
def adicionar_elementos(lista):
    # adicionando 2 elementos
    lista.append("novo1")
    lista.append("novo2")

    # imprimindo a lista atualizada
    print(lista)

# chamada da função
minha_lista = [1, 2, 3, 4]
adicionar_elementos(minha_lista)
```

## Exercício 4.

```python
def mostrar_elementos(elementos):
    print("Elementos:", elementos)

# chamada com 1 elemento
mostrar_elementos([10])

# chamada com 4 elementos
mostrar_elementos([10, 20, 30, 40])
```
