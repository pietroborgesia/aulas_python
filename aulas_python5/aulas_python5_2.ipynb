## Exercício 6

```python
Celsius = [39.2, 36.5, 37.3, 37.8]

converter = lambda lista: list(map(lambda c: round((9/5)*c + 32, 2), lista))

Fahrenheit = converter(Celsius)
```

```python
Fahrenheit
```

## Exercício 7

```python
def amplitude(lista):
    return max(lista) - min(lista)
```

```python
valores = [10, 5, 8, 20, 3]

resultado = amplitude(valores)
```

```python
resultado
```

## Exercício 8

```python
def sequencia(n):
    for i in range(1, n + 1):
        print((str(i) + " ") * i)
```

```python
sequencia(8)
```
