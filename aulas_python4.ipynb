### Exemplo de "and"

```python
idade_pessoa = int(input("Digite a sua idade:"))
tem_documento = input("Tem documento (s/n)")

if idade_pessoa >= 18 and tem_documento == "s":
    print("Entrada permitida")
else:
    print("Não pode entrar!")
```

### Exemplo de "or"

```python
idade_pessoa = int(input("Digite a sua idade:"))
estudante = input("É estudante? (s/n)")

if idade_pessoa <= 12 or estudante == "s":
    print("Entrada permitida")
else:
    print("Não pode entrar!")
```

## Exemplo com "not"

## While

```python
contador = 1
while contador <= 5:
    print(contador)
    contador = contador + 1
```

```python
contador = 10
while contador <= 5:
    print(contador)
    contador -= 1
```

```python
numero_atual = 0
while numero_atual <= 10:
    print(numero_atual)
    numero_atual = numero_atual + 1
```

## MENU

```python
opcao = " "
while opcao != "0":
    opcao = input("Digite um número ou 0 para sair: ")
    print(opcao)
    print ("1. Opção 1")
    print ("2. Opção 2")
    print ("0 - Sair ")
```

## Range

```python
for numero in range(6):
  print(numero, end= ",")
```

```python
#exemplo progressão aritmética 3 em 3
for numero in range(0, 29, 3):
  print(numero, end=",")
```
