Hora de praticar!
Crie um programa que:
1. Peça a idade do usuário.
2. Mostre na tela:
▶ Criança (se for menor de 13 anos)
▶ Adolescente (de 13 a 17 anos)
▶ Adulto (18 anos ou mais)

```python
idade = int(input("Digite sua idade:"))

if idade < 13:
    print("Criança")
elif 13<= idade <= 17:
    print("Adolescente")
else:
    print("Adulto")
```

Exercícios Práticos – Nível Fácil

1 Verificar número positivo
Peça ao usuário um número e verifique: Se o número for
maior que 0, mostre "Número positivo". (Use apenas
if).

```python
numero = int(input("Digite um número:"))

if numero >=0:
    print("Número positivo", numero)
```

2 Verificar maior de idade
Peça a idade do usuário. Mostre "Maior de idade" se a
idade for maior ou igual a 18. (Use apenas if).

```python
idade = int(input("Digite sua idade:"))

if idade >=18:
    print("Maior de idade", idade)
```

3. Número maior que 10
Peça um número ao usuário. Se o número for maior que
10, mostre "Número maior que 10". (Use apenas if).

```python
numero = int(input("Digite um número maior que 10:"))

if numero >=10:
    print("Número maior que 10.")
```

4- Par ou Ímpar
Peça um número inteiro. Mostre "Número par" ou
"Número ímpar". (Use if e else).

```python
numero = int(input("Digite um número inteiro:"))

if numero %2 == 0:
    print("Número par.")

else:
    print("Número impar.")
```

5- Aprovação de aluno
Peça a nota do aluno. Se a nota for maior ou igual a 7,
mostre "Aluno aprovado". Caso contrário, mostre "Aluno
reprovado". (Use if e else).

```python
nota = int(input("Digite sua nota:"))

if nota >=7:
    print("Aluno aprovado.")

else:
    print("Aluno reprovado.")
```

6- Comparar dois números
Peça dois números ao usuário. Mostre qual é maior.
Exemplo: "O maior número é 25". (Use if e else).

```python
numero1 = int(input("Digite um número:"))
numero2 = int(input("Digite um outro número:"))

if numero1 > numero2:
    print("Primeiro número maior.", numero1)

else:
    print("Este número é maior", numero2)
```

8- Classificação de idade
Peça a idade da pessoa. Mostre:
▶ Criança (0–12)
▶ Adolescente (13–17)
▶ Adulto (18 ou mais)
(Use if, elif e else).

```python
idade = int(input("Digite sua idade:"))

if idade < 13:
    print("Criança")
elif 13<= idade <=17:
    print("Adolescente.")
else:
    print("Adulto")
          
```

8- Nota do aluno
Peça a nota do aluno. Mostre:
▶ Aprovado (nota ≥ 7)
▶ Recuperação (nota entre 5 e 6)
▶ Reprovado (nota < 5)
(Use if, elif e else).

```python
nota = int(input("Digite sua nota:"))

if nota >=7:
    print("Aluno aprovado.")

elif nota >= 5:
    print("Aluno recuperação.")

else:
    print("Reprovado.")
    
        
```

9- Comparar três números
Peça três números. Mostre qual é o maior deles.
Exemplo: "O maior número é 42".

```python
numero1 = int(input("Digite primeiro número:"))
numero2 = int(input("Digite segundo número:"))
numero3 = int(input("Digite terceiro número:"))

if numero1 > numero2 and numero1 > numero3:
    print("Primeiro número maior.")

elif numero2 > numero1 and numero2 > numero3:
    print("Segundo número é maior.")

else:
    print("Terceiro número maior.")
```

Crie um programa que peça um número e mostre:
▶ Número positivo
▶ Número negativo
▶ Número igual a zero

```python
numero = int(input("Digite um número:"))

if numero <0:
    print("Número negativo", numero)
elif numero >0:
    print("Número positivo", numero)
else:
    print("Número igual a zero")
```
