## Exercícios C - FOR

```python
#exercicio 1
#Mostre os números de 1 até 5.

for numero in range(1, 6):
  print(numero, end = ",")
```

```python
#exercicio 2
#Mostre os números pares de 0 até 20.

for numero in range(0, 22, 2):
  print(numero, end ="'")
```

```python
#exercicio 3.
#Mostre a tabuada do 5.

for numero in range(1,11):
  print(f"5 x{numero} =", 5*numero)
```

```python
#exercicio 4.
#Repita a frase "Python é incrível" 4 vezes.

for x in range(5):
  print("Python é incrivel")
```

```python
#exercicio 5.
#Percorra a palavra "Pernambuco" e mostre cada letra.

for letra in "Pernambuco":
  print(letra, end = " \n")
```

```python
#exercicio 6.
#Some os números de 1 até 10 usando for.

soma = 0

for numero in range(1,11):
  soma +=numero
  print(soma)
```

## Exercícios D — Mistos

```python
#exercicio 1.
#Peça 5 números com for e diga quantos são pares.

pares = 0

for numero in range(5):
    valor = int(input("Digite um número: "))

    if valor % 2 == 0:
        pares += 1

print("Números pares são:", pares)
```

```python
#exercicio 2.
#Use while para pedir uma nota entre 0 e 10.

nota = -1

while nota < 0 or nota > 10:
    nota = int(input("Digite uma nota entre 0 e 10: "))

print("Nota válida:", nota)
```

```python
#exercicio 3.
#Peça um número e mostre a tabuada com for.

numero = int(input("Digite um número: "))

for tabuada in range(1, 11):
    print(numero, "x", tabuada, "=", numero * tabuada)
```

```python
#exercicio 4.
#Peça idade e documento. Use if com and para verificar acesso.


idade = int(input("Digite sua idade: "))
documento = (input("Você tem documento? (sim/não): "))

if idade >= 18 and documento == "sim":
    print("Acesso permitido.")
else:
    print("Acesso negado.")
```

```python
#exercicio 5.
#Peça uma palavra e mostre letra por letra com for.


letra = input("Digite uma palavra: ")

for caracteres in letra:
  print(caracteres, end = " \n")
```

```python
#exercicio 6.
#Crie um sistema de login com 3 tentativas usando while.


senha_correta = "1234"
usuario_correto = "admin"

tentativa = 0
limite = 3

while tentativa < limite:
  usuario = input("Digite o usuário: ")
  senha = input("Digite a senha: ")


  if usuario == usuario_correto and senha == senha_correta:
    print("Login realizado com sucesso!")
    break

  else:
    tentativa += 1
  print(f"Incorreto, tente novamente{tentativa, limite}")

  if tentativa == limite:
    print("Acesso bloqueado!")
```
