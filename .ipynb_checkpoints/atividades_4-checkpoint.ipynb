## Exercícios A — Operadores Lógicos.

1. Peça a idade e se a pessoa tem carteira de motorista.
Mostre se ela pode dirigir.

```python
idade = int(input("Digite sua idade."))
cnh = input("Possui habilitação? (S/N)")

if idade >= 18 and cnh == "S":
    print("Pode dirigir!")

else:
    print("Não pode dirigir!")
```

2. Peça um número. Verifique se ele está entre 50 e 100.

```python
numero = int(input("Digite um número da sorte."))

if numero >= 50 and numero <= 100:
    print("Acertou!")

else:
    print("ERRRROOOUUUUU, tente novamente!")
```

3. Peça o dia da semana. Mostre se é final de semana.

```python
dia = input("Digite o dia da semana: ")

if dia != "sabado" and dia != "domingo":
    print("Não é final de semana.")
else:
    print("É final de semana.")
```

4. Peça usuário e senha. Verifique se ambos estão corretos.

```python
usuario = input("Digite o usuário: ")
senha = input("Digite a senha: ")

if usuario == "admin" and senha == "1234":
    print("Login correto!")
else:
    print("Usuário ou senha incorretos.")
```

5. Peça uma nota e a frequência do aluno. Mostre se ele foi
aprovado.

```python
nota = int(input("Digite sua nota: "))
frequencia = int(input("Digite sua frequência: "))

if nota >= 6 and frequencia >= 75:
    print("Aprovado!")
else:
    print("REPROVADO!")
```

6. Peça se o aluno entregou o trabalho. Use not para verificar
se ele não entregou.

```python
entregou = input("Você aluno entregou o trabalho do Ujeverson?: ")

if not entregou == "sim":
    print("Aluno não entregou o trabalho.")
else:
    print("Aluno entregou o trabalho.")
```

## Exercícios B — while

1. Mostre os números de 1 até 10 usando while.

```python
numero = 1

while numero <= 10:
    print(numero)
    numero = numero + 1
```

2. Peça uma senha até o usuário acertar.

```python
senha_correta = "1234"
senha = ""

while senha != senha_correta:
    senha = input("Digite a senha: ")
print("Senha correta!")
```

3. Peça números até o usuário digitar 0.

```python
numero_correto = "0"
numero = ""

while numero != numero_correto:
    numero = input("Digite um número até acertar: ")
print("Acertoouuu!!")
```

4. Conte de 10 até 1.

```python
numero = 10

while numero >= 1:
    print(numero)
    numero = numero - 1
```

5. Peça uma idade válida até o usuário digitar um valor positivo.

```python
idade = -1

while idade <= 0:
    idade = int(input("Digite sua idade: "))
print("Você está vivo.")
```

6. Crie um menu que só encerra quando o usuário digitar 0.

```python
opcao = -1

while opcao != 0:
    print("\n1 - Opção 1")
    print("2 - Opção 2")
    print("3 - Opção 3")
    print("0 - Sair")

    opcao = int(input("Escolha uma opção: "))

    if opcao == 1:
        print("Você escolheu a opção 1")
    elif opcao == 2:
        print("Você escolheu a opção 2")
    elif opcao == 3:
        print("Você escolheu a opção 3")
    elif opcao == 0:
        print("Saindo do programa...")
    else:
        print("Opção inválida.")
```
