### Exemplo de "and"

```python
idade_pessoa = int(input("Digite a sua idade:"))
tem_documento = input("Tem documento (s/n)")

if idade_pessoa >= 18 and tem_documento == "s":
    print("Entrada permitida")
else:
    print("Não pode entrar!")
```

### Exemplo de "or"

```python
idade_pessoa = int(input("Digite a sua idade:"))
estudante = input("É estudante? (s/n)")

if idade_pessoa <= 12 or estudante == "s":
    print("Entrada permitida")
else:
    print("Não pode entrar!")
```

## Exemplo com "not"

```python
senha_correta = False
if not senha_correta:
    print("Senha incorreta!")
else:
    print("Senha correta!")
```

## Número entre 10 e 20 (como casas de apostas funcionam.)

```python
numero = int(input("Digite um número!"))
if numero >= 10 and numero <= 20:
    print("Acertou!")
else:
    print("Tente novamente!")
```

## Dia Semana

```python
dia_semana = input("Digite o dia da semana:")

if dia_semana == "sabado" or dia_semana == "Domingo":
    print("Fim de semana")
else:
    print("Ainda estão na escala 6x1!")
```
