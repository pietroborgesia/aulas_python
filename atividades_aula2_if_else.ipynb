1- Acesso ao parque

```python
idade = int(input("Digite sua idade: "))
ingresso = input("Tem ingresso? (sim/não): ")

if idade >= 12:
    if ingresso == "sim":
        print("Entrada permitida")
    else:
        print("Precisa comprar ingresso")
else:
    print("Idade insuficiente")
```

2- Login Simples

```python
nome_do_usuario = input("Digite o usuário: ")

if nome_do_usuario == "admin":
    senha_do_usuario = input("Digite a senha: ")
    
    if senha_do_usuario == "1234":
        print("Acesso permitido")
    else:
        print("Senha incorreta")
else:
    print("Usuário inválido")
```

3- Aprovação com presença

```python
nota_aluno = float(input("Digite a nota: "))

if nota_aluno >= 7:
    frequencia_aluno = float(input("Digite a frequência: "))
    
    if frequencia_aluno >= 75:
        print("Aprovado")
    else:
        print("Reprovado por falta")
else:
    print("Reprovado por nota")
```

4- Número par positivo

```python
numero_di = int(input("Digite um número inteiro: "))

if numero_di > 0:
    if numero_di % 2 == 0:
        print("Par positivo")
    else:
        print("Ímpar positivo")
else:
    print("Número não positivo")
```

5- Entrada no cinema

```python
idade_da_pessoa = int(input("Digite a idade: "))

if idade_da_pessoa >= 16:
    possui_documento = input("Possui documento? (s/n): ")
    
    if possui_documento == "s":
        print("Pode entrar")
    else:
        print("Precisa apresentar documento")
else:
    print("Entrada não permitida")
```

6- Desconto em compra

```python
compra = float(input("Digite o valor da compra: "))

if compra >= 100:
    cupom = input("Possui cupom? (s/n): ")
    
    if cupom == "s":
        print("Desconto aplicado")
    else:
        print("Sem desconto")
else:
    print("Valor insuficiente para desconto")
```

7- Múltiplo de 2 e de 3

```python
numero = int(input("Digite um número inteiro: "))

if numero > 0:
    if numero % 2 == 0:
        if numero % 3 == 0:
            print("Positivo, múltiplo de 2 e de 3")
        else:
            print("Positivo e múltiplo de 2, mas não de 3")
    else:
        print("Positivo, mas não múltiplo de 2")
else:
    print("Número não positivo")
```

8- Cadastro com confirmação

```python
nome_da_pessoa = input("Digite o nome: ")

if nome_da_pessoa != "":
    idade_da_pessoa = int(input("Digite a idade: "))
    
    if idade_da_pessoa >= 18:
        print("Cadastro completo")
    else:
        print("Cadastro incompleto: menor de idade")
else:
    print("Nome obrigatório")
```

9- Triagem hospitalar simples

```python
idade_pessoa = int(input("Digite a idade: "))

if idade_pessoa >= 60:
    dor_forte = input("Sente dor forte? (s/n): ")
    
    if dor_forte == "s":
        print("Atendimento prioritário imediato")
    else:
        print("Aguardar triagem")
else:
    print("Atendimento comum")
```

10- Sistema de acesso por níveis

```python
usuario_do_sistema = input("Digite o usuário: ")

if usuario_do_sistema == "professor":
    senha_do_sistema = input("Digite a senha: ")
    
    if senha_do_sistema == "python123":
        nivel_de_acesso = input("Digite o nível de acesso: ")
        
        if nivel_de_acesso == "1":
            print("Acesso básico")
        else:
            print("Acesso avançado")
    else:
        print("Senha inválida")
else:
    print("Usuário não autorizado")
```
